```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import json

# Step 1: Load the transition matrix JSON file
transition_matrix_path = "C:/Users/Gsgaa/OneDrive/Desktop/Markov Trading Bot/data/multidimensional_transition_matrix_AAPL.json"  # Update the path as needed
with open(transition_matrix_path, "r") as f:
    transition_matrix = json.load(f)

# Step 2: Convert the JSON matrix into a DataFrame
matrix_df = pd.DataFrame(transition_matrix).fillna(0)

# Step 3: Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(matrix_df, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("State Transition Matrix Heatmap")
plt.xlabel("Next State")
plt.ylabel("Current State")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()

# Step 4: Show the plot
plt.show()
```
